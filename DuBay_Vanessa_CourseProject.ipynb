```python
import sys
file = open("Salt_Marsh_data.txt")
store_data = [] #Initializing a list to store all the data once the file is read through.
for line in file:
    if line[0]=="#": #Determines if the data is a header. 
        header = line.split('\t') 
    else:
        data = line.split('\t') #Separates the data from the headers.
        store_data.append(data) #Stores the data to the initialized store_data list.
 
#Initializing lists of data to store while going through the data. 
juvenile_crab = []
juvenile_crab_biomass = []
snails = []
snails_biomass = []
site_list = []

#Grabbing of data
for samples in store_data:
    site = samples[0]
    site_list.append(site)
    juvenilecrab = samples[6]
    juvenile_crab.append(float(juvenilecrab)) #Making sure each value grabbed is a float value to use in calculations. 
    juvenilecrabbiomass = samples[7]
    juvenile_crab_biomass.append(float(juvenilecrabbiomass))
    snail = samples[4]
    snails.append(float(snail))
    snailbiomass = samples[5]
    snails_biomass.append(float(snailbiomass))

#Changing site data to a numeric value alphabetically using a dictionary. 
numerical_site_dictionary = {'AM': 1, 'BF': 2, 'BL': 3, 'BR': 4, 'BW': 5, 'JI': 6, 'LS': 7, 'PL': 8, 'SS': 9}

numerical_site = [numerical_site_dictionary[s] for s in site_list]

#Mean, Median, Standard Deviation Calculations using Numpy functions
import numpy as np
site_mean = round(np.mean(numerical_site),2) #Rounding to second decimal place for visual appeal. 
site_median = round(np.median(numerical_site),2)
site_std = round(np.std(numerical_site),2)
print(f"The mean of the site variable (in numeric terms) is {site_mean}, the median is {site_median}, and the standard deviation is {site_std}.")

juvenile_crab_mean = round(np.mean(juvenile_crab),2)
juvenile_crab_median = round(np.median(juvenile_crab),2)
juvenile_crab_std = round(np.std(juvenile_crab),2)
print(f"The mean of the juvenile fiddler crab variable is {juvenile_crab_mean}, the median is {juvenile_crab_median}, and the standard deviation is {juvenile_crab_std}.")

juvenile_crab_biomass_mean = round(np.mean(juvenile_crab_biomass),2)
juvenile_crab_biomass_median = round(np.median(juvenile_crab_biomass),2)
juvenile_crab_biomass_std = round(np.std(juvenile_crab_biomass),2)
print(f"The mean of the juvenile fiddler crab biomass variable is {juvenile_crab_biomass_mean}, the median is {juvenile_crab_biomass_median}, and the standard deviation is {juvenile_crab_biomass_std}.")

snails_mean = round(np.mean(snails),2)
snails_median = round(np.median(snails),2)
snails_std = round(np.std(snails),2)
print(f"The mean of the snails variable is {snails_mean}, the median is {snails_median}, and the standard deviation is {snails_std}.")

snails_biomass_mean = round(np.mean(snails_biomass),2)
snails_biomass_median = round(np.median(snails_biomass),2)
snails_biomass_std = round(np.std(snails_biomass),2)
print(f"The mean of the snails biomass variable is {snails_biomass_mean}, the median is {snails_biomass_median}, and the standard deviation is {snails_biomass_std}.")

#Creating histograms of each variable used.
import matplotlib.pyplot as plt

plt.hist(numerical_site, bins=9, edgecolor = 'black') #Adding an edge color to distinguish the lines of the bins and using 9 bins to show amount of sites and how frequently they appear.  
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram of Numerical Site Values')
plt.show()

plt.hist(juvenile_crab, bins=11, edgecolor = 'black')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram of Juvenile Fiddler Crab Values')
plt.show()

plt.hist(juvenile_crab_biomass, bins=11, edgecolor = 'black')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram of Juvenile Fiddler Crab Biomass Values')
plt.show()

plt.hist(snails, bins=11, edgecolor = 'black')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram of Snails Values')
plt.show()

plt.hist(snails_biomass, bins=11, edgecolor = 'black')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.title('Histogram of Snails Biomass Values')
plt.show()
```

```python
#Imports
from scipy import stats
from scipy.stats import linregress

#Pearson and Spearman Coefficient Correlation Statistic Calculations for the Juvenile Fiddler Crabs
PCC = stats.pearsonr(numerical_site,juvenile_crab)
SCC = stats.spearmanr(numerical_site, juvenile_crab)
print("Juvenile Fiddler Crab Pearson Coefficient =", PCC.statistic)
print("Juvenile Fiddler Crab Spearman Coefficient =", SCC.statistic)

#Linear regression line calculation
x = np.array(numerical_site)
y = np.array(juvenile_crab)
slope, intercept, r_value, p_value, std_err = linregress(x, y)
regression_line = slope * x + intercept

#Creating scatterplot of data and plotting regression line on the scatterplot. 
plt.scatter(x, y, color='green', label='Juvenile Fiddler Crab Data')
plt.plot(x, regression_line, color='red', label='Regression Line')
plt.xlabel('Sample Site')
plt.ylabel('Juvenile Fiddler Crab Abundance')
plt.title('Juvenile Fiddler Crab vs Site')
plt.legend()
plt.show()

#Calculation of Pearson and Spearman Coefficients for the Snails
PCC = stats.pearsonr(numerical_site,snails)
SCC = stats.spearmanr(numerical_site, snails)
print("Snails Pearson Coefficient =", PCC.statistic)
print("Snails Spearman Coefficient =", SCC.statistic)

#Linear regression line calculation
x2 = np.array(numerical_site)
y2 = np.array(snails)
slope, intercept, r_value, p_value, std_err = linregress(x2, y2)
regression_line = slope * x2 + intercept
plt.scatter(x2, y2, color='blue', label='Snail Data')
plt.plot(x, regression_line, color='orange', label='Regression Line')
plt.xlabel('Sample Site')
plt.ylabel('Snails Abundance')
plt.title('Snails vs Site')
plt.legend()
plt.show()
```

```python
#Imports
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

#Storing data that is numerical and impacts the overall data (from column 5 to column 16)
numerical_data = []
for ii in store_data:
    numerical_data.append(ii[5:16])

#Converting this data to a float value but still keeping it as a list of lists. 
totaldata = []
for jj in numerical_data:
    row_by_row = []
    for xx in jj:
        row_by_row.append(float(xx))
    totaldata.append(row_by_row)

#Using numpy to make an array and perform calculations to assist in PCA analysis (standardizing the data). 
total_numeric_array = np.array(totaldata)
numeric_means = np.mean(total_numeric_array, axis = 0)
numeric_std = np.std(total_numeric_array, axis = 0)
standardized_data = (total_numeric_array - numeric_means) / numeric_std

#PCA data and components to use in bar plot. 
pca = PCA()
pca_results = pca.fit_transform(standardized_data)
components = pca.components_
explained_variance = pca.explained_variance_ratio_

#Creating bar plot of the weight components of each principal component found in the data. 
plt.figure(figsize = (12,6))
plt.bar(range(1,len(explained_variance)+1), explained_variance*100, color = 'blue')

#Adding labels and title.
plt.xlabel('Principal Component')
plt.ylabel('Percentage of Variance')
plt.title('Weight Components of Principal Components (in percentage)')
plt.xticks(range(1, len(explained_variance) + 1)) #Adding individual x tick marks for clarity. 
plt.grid(axis='y') #Plotting a grid only on the y-axis just for further readability. 
plt.show()  
print("Explained Variance Ratio for all components:",pca.explained_variance_ratio_)

#Creating second PCA analysis on the two components of PC1 and PC2
pca2 = PCA(n_components=2)
pca_scores2 = pca2.fit_transform(standardized_data)
plt.figure(figsize = (8,6))
plt.scatter(pca_scores2[:,0], pca_scores2[:,1], color = 'skyblue')

#Creating a dictionary and assigning colors to each site and using a for loop to apply the color to the site, but not repeating the site in the legend while having it keep the same color. 
site_colors = {'AM': 'red', 'BF': 'green', 'BL': 'black', 'BR': 'orange', 'BW': 'purple', 'JI': 'teal', 'LS': 'pink', 'PL': 'yellow', 'SS': 'violet'}
colors = [site_colors[site] for site in site_list]
plotted_site = set() #Empty set to add already plotted sites to.
for ii in range(len(site_list)):
    site = site_list[ii]
    if site not in plotted_site:
        plotted_site.add(site) 
        plt.scatter(pca_scores2[ii, 0], pca_scores2[ii, 1], c=colors[ii], label=site_list[ii])
#Adding the snails abundance and juvenile fiddler crabs abundance data to the PCA graph and seeing where it overlaps with the site data as well as the total PCA variance data. Using different markers so it is easily distinguishable. 
plt.scatter(pca_scores2[:, 0], pca_scores2[:, 1], c=snails, marker='|', label='Snail Abundance')
plt.scatter(pca_scores2[:, 0], pca_scores2[:, 1], c=juvenile_crab, marker='_', label='Juvenile Crabs Abundance')

#Adding labels and title
plt.xlabel(f'PC1 ({pca2.explained_variance_ratio_[0] * 100:.2f}% Variance)')
plt.ylabel(f'PC2 ({pca2.explained_variance_ratio_[1] * 100:.2f}% Variance)')
plt.title('PC1 vs PC2')
plt.legend()
plt.grid()
plt.show()

print("Explained Variance Ratio for PC1 and PC2:",pca2.explained_variance_ratio_)
```
